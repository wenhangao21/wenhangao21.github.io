## Cool Examples

Colored mandelbrot set:

```python
import matplotlib.pyplot as plt
import numpy as np

def mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iterations):
    # Create an empty image to store the Mandelbrot Set visualization
    image = np.zeros((height, width))

    # Calculate the pixel size based on the provided bounds and image size
    x_range = np.linspace(xmin, xmax, width)
    y_range = np.linspace(ymin, ymax, height)

    # Calculate the Mandelbrot Set for each pixel
    for x_index, x in enumerate(x_range):
        for y_index, y in enumerate(y_range):
            c = complex(x, y)
            z = 0
            iterations = 0

            while abs(z) <= 2 and iterations < max_iterations:
                z = z*z + c
                iterations += 1

            # Map the number of iterations to a color value
            color = iterations % 256

            # Set the color value for the corresponding pixel in the image
            image[y_index, x_index] = color

    return image

def main():
    # Specify the image size and Mandelbrot Set region of interest
    width = 800
    height = 600
    xmin, xmax = -2.0, 1.0
    ymin, ymax = -1.5, 1.5
    max_iterations = 1000

    # Generate the Mandelbrot Set image
    mandelbrot_image = mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iterations)

    # Display the image using matplotlib
    plt.imshow(mandelbrot_image, cmap='hot', extent=(xmin, xmax, ymin, ymax))
    plt.title('Mandelbrot Set Visualization')
    plt.xlabel('Re(c)')
    plt.ylabel('Im(c)')
    plt.colorbar(label='Iterations')
    plt.show()

if __name__ == "__main__":
    main()
```

# **Lecture 5 Part 2 Linear Regression with Numpy and Visualization with Matplotlib**

## Intro to Linear Regression

What is Linear Regression?
* Linear regression is a statistical method used for modeling the relationship between a dependent variable (target) and one or more independent variables (features).
* It aims to find a linear equation that best fits the data.

Key Applications

* Prediction: Linear regression can be used for predicting numerical values (e.g., predicting house prices based on features like square footage and number of bedrooms).
* Understanding Relationships: It helps understand how changes in independent variables affect the dependent variable.

 Linear Regression involves finding the best linear hypothesis (a line or a hyperplane) on the data. The equation of a linear relationship is in the format of $y = w^Tx + b$ where w is the slope and b is the intercept. In machine learning terms, we call $w$ the weight(s) and $b$ the bias(es).

Optimization Problem:
* A set of $m$ points (known data points) $(x^{(i)}, y^{(i)})$ is given.
* We were to find the values of $w$ and $b$ that minimizes the mean square error:
$$(w,b)^* = \underset{w, b}{\operatorname{arg min}} J\left(w, b\right) = \frac{1}{m} \sum_{i=0}^m\left(y^{(i)}-\left(w^T x^{(i)}+b\right)\right)^2$$

## Finding Parameter Values with Linear Algebra

We are going to use some Linear Algebra concepts for finding the right weight and bias terms for our data.

We can write each data point $(x^{(i)}, y^{(i)})$ as
$$
\begin{array}{ccc}
w^Tx^{(1)}+b=y^{(1)} & \text { for } & (x^{(1)},y^{(1)}) \\
w^Tx^{(2)}+b=y^{(2)} & \text { for } & (x^{(2)},y^{(2)}) \\
\cdots&
\end{array}
$$

If we use $\Theta$ to denote $\left[\begin{array}{l}w \\b\end{array}\right]$, we can rewrite the system of equations as:
$$X \Theta=Y$$
$$X^T X \Theta=X^T Y$$
Assuming $\left(\mathbf{X}^{\top} \mathbf{X}\right)$ to be invertible.
$$\Theta=\left(X^T X\right)^{-1} X^T Y$$
Now, we have derived a formula to find $\theta$. This is known as the Normal equation.

## Implementation with Numpy and Visualization with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # Random feature values between 0 and 2
y = 3 * X + 1 + np.random.randn(100, 1)  # Linear relationship with noise

# Plot the data
plt.scatter(X, y, label='Data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic Data for Linear Regression')
plt.legend()
plt.show()
plt.close
```

```python
# Add the coefficients of bias terms to X, all 1s
X_b = np.c_[X, np.ones((100, 1))]
print(X_b[0:5])
# Calculate the optimal coefficients using the normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Get the coefficients
w = theta_best[0]
b = theta_best[1]

# Make predictions on new data
X_new = np.array([[1.5]])  # New data point
y_pred = w*X_new + b

# Plot the data and regression line
plt.scatter(X, y, label='Data')
plt.plot(X, X_b.dot(theta_best), color='red', linewidth=2, label='Regression Line')
plt.scatter(X_new, y_pred, color='green', marker='x', s=100, label='Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression Example')
plt.show()

# Print the model parameters
print(f"Slope (Coefficient): {w}")
print(f"Intercept: {b}")
```

# **Lecture 5 Part 3 Logistic Regression with Numpy and Visualization with Matplotlib**

## Introduction to Classification


Binary Classification

* Binary classification is a type of supervised learning problem where the goal is to classify input data into one of two classes (e.g., yes/no, spam/not spam, healthy/sick).

The Need for Logistic Regression

* Linear regression, which is used for predicting continuous values, is not suitable for classification tasks. Logistic regression, on the other hand, is designed specifically for binary classification and provides a probability estimate that an input belongs to a particular class.

## Logistic (Sigmoid) Function

The sigmoid function, denoted as (z), is a key component of logistic regression. It maps any real-valued number to a value between 0 and 1, making it suitable for modeling probabilities.
$$
\sigma(x)=\frac{1}{1+e^{-x}}=\frac{e^x}{1+e^x}=1-\sigma(-x) .
$$
![sigmoid](https://www.baeldung.com/wp-content/uploads/sites/4/2022/02/sigmoid.png)

## Linear Regression vs. Logistic Regression

Differences in Objective

* Linear regression aims to predict a continuous output.
* Logistic regression models the probability of an input belonging to a class.

Output Interpretation

* In linear regression, the output is a continuous value.
* In logistic regression, the output is the probability of belonging to a specific class.

## Hypothesis Function

The logistic regression hypothesis function $y_{\theta}(x)$ is defined as:
$$y_{\theta}(x) = \sigma(w^T x +b),$$
here $\theta$ denotes that parameters $w$ and $b$.

## The Cost Function

The cost function in logistic regression is commonly referred to as the log loss or cross-entropy loss. It measures the dissimilarity between the predicted probabilities and the actual class labels.

Here, the loss function with respect to a single data point is given as:
$$
\mathcal{L}\left(a^{(i)}, y^{(i)}; \theta \right)=-y^{(i)} \log \left(a^{(i)}\right)-\left(1-y^{(i)}\right) \log \left(1-a^{(i)}\right),
$$
where $a^{(i)} = y_\theta\left(x^{(i)}\right)$ is the prediction probability.

**Interpreting the Components:** In the case of binary classification, the loss function can be decomposed into two parts:
- Log Loss for Positives: The term $-y^{(i)} \log \left(a^{(i)}\right)$ in the loss function penalizes the model when it predicts a probability close to 0 for positive examples.
  - When the predicted probability is closer to $0$, $-\log \left(a^{(i)}\right)$ is going to be larger, the loss will be larger, in a way that penalizes the model more.
- Log Loss for Negatives: The term $-\left(1-y^{(i)}\right) \log \left(1-a^{(i)}\right)$ in the loss function penalizes the model when it predicts a probability close to 0 for negative examples.
  - - When the predicted probability is closer to $1$, $-\log \left(1- a^{(i)}\right)$ is going to be larger, the loss will be larger, in a way that penalizes the model.

The cost is then computed by averaging over all training examples:
$$J(\theta)=\frac{1}{m} \sum_{i=1}^m \mathcal{L}\left(a^{(i)}, y^{(i)}\right)$$

**Fact:** This (cross-entropy) loss function is conveniently **convex**.




The goal is to find the parameter values that minimize this cost function.
$$\theta^*=\underset{\theta}{\arg \min }~J (\boldsymbol{\theta})$$

## Vectorization

Here, we write $X$, $Y$, $A$ to denote the vectors representing $x^{(i)}$, $y^{(i)}$, and $a^{(i)}$. For $X$, we use the convention that $x^{(i)}$-s are stacked horizontally, so the shape is $(d, m)$, but you can do the other way, just a matter of reshaping/tranpose/change dot product correspondingly; for $Y$ and $A$, they are now 1D arrays.





Then, we have
$$A=\sigma\left(w^T X+b\right),$$
and
$$J=-\frac{1}{m} \sum \Big[Y * \log \left(A\right)+\left(1-Y\right) * \log \left(1-A\right)\Big],$$
where all operations (addition, subtraction, multiplication) are point-wise, and the summation just sums over all the elements in the vector.

## Gradient Descent

Now, we have the cost function, but how do we find the parameters that minimize this cost function? The answer is through gradient descent.

In gradient descent, our objective is to solve the optimization problem: $\min_{x\in \mathbb{R}^d}\; f(x)$, where $f$ is a convex function $f: \mathbb{R}^n \rightarrow \mathbb{R}$.

Algorithm description:
- Start with some initial guess, $x_0$.
- Generate new guess $x_1$ by moving in the negative gradient direction:
$$
x_1=x_0-\alpha_0 \nabla f\left(x_0\right),
$$
where $\alpha_0$ is the step size.
- Repeat to successively refine the guess:
$$
x_{k+1}=x_{k}-\alpha_k \nabla f\left(x_k\right), \quad \text { for } k=1,2,3, \ldots
$$
where we might use a different step-size $\alpha_k$ on each iteration.
- Stop when stopping criteria is met.

- In practice, you can stop if you detect that you aren't making progress, or if $\left\|\nabla f\left(w^k\right)\right\| \leq \epsilon$, or if a certain number of iterations has been conducted.

In our logistic regression cost function setting, the updates are
$$
\theta_{k+1}=\theta_{k}-\alpha_{k} \nabla J\left(\theta_{k}\right).
$$

## Convergence of Gradient Descent

**Assumption (Lipschitz Gradients):**

>The function $f: \mathbb{R}^n \rightarrow \mathbb{R}$ has L-Lipschitz continuous gradients.

>In other words, there exists a positive constant $L$ such that $$\left\|\nabla f(v)-\nabla f(w)\right\| \leq L\|v-w\|,\;\;\; \forall v,w\in \mathbb{R}^d$$

What this assumption suggests is that Gradients cannot change arbitrarily fast.

**Lemma (Descent Lemma):**

> If $f: \mathbb{R}^n \rightarrow \mathbb{R}$ has L-Lipschitz continuous gradients, then $$
f(v) \leq f(w)+\nabla f(w)^T(v-w)+\frac{L}{2}\|v-w\|^2,\;\;\; \forall v,w\in \mathbb{R}^d$$

The descent lemma gives us a convex quadratic upper bound on $f$.

A proof to this lemma can be found at "Zhou, Xingyu. On the Fenchel Duality between Strong Convexity and Lipschitz Continuous Gradient. [arXiv preprint](https://arxiv.org/abs/1803.06573) arXiv:1803.06573 (2018)". It is fairly simple, as long as you have taken courses in linear algebra and calculus, you should be able to understand the proof.

**Theorem:**

> If $f: \mathbb{R}^n \rightarrow \mathbb{R}$ has L-Lipschitz continuous gradients, and $f^*=\min _x f(x)>-\infty$, then the gradient descent algorithm with fixed step size satisfying $\alpha <\frac{2}{L}$ will converge to a stationary point.


Here we assume a fixed step size and $\alpha <\frac{2}{L}$ only for the purpose of a mathematical proof. In practice, you should never use $\alpha<\frac2L$. $L$ is usually expensive to compute, and this step-size is really small. You only need a step-size this small in the worst case. Usually, you can start with a larger step size and gradually decrease it.


Note that the Descent Lemma also holds for $C^1$ functions, you can find more information online if you are interested.

**Proof of the Theorem:**

Sketch of the proof: we show that $f$ is always decreasing in every iteration; that is $f\left(x_{k+1}\right) \leq f\left(x_{k}\right) - \xi$ for some $\xi >0$.

Recall the Descent Lemma:
$$
f(v) \leq f(w)+\nabla f(w)^T(v-w)+\frac{L}{2}\|v-w\|^2
$$

If we substitle $x_{k+1}$ and $x_k$ into the descent lemma we get:
$$
f\left(x_{k + 1}\right) \leq f\left(x_k\right)+\nabla f\left(x_k\right)^T\left(x_{k+1}-x_k\right)+\frac{L}{2}\left\|x_{k + 1}-x_k\right\|^2
$$
Now if we use that $\left(x_{k+1}-x_k\right)=-\alpha\nabla f\left(x_k\right)$ in gradient descent:
$$
\begin{aligned}
f\left(x_{k+1}\right) & \leq f\left(x_k\right)-\alpha \nabla f\left(x_k\right)^T \nabla f\left(x_k\right)+\frac{\alpha L}{2}\left\|\alpha \nabla f\left(x_k\right)\right\|^2 \\
& =f\left(x_k\right)-\alpha\left\|\nabla f\left(x_k\right)\right\|^2+\frac{\alpha^2 L}{2}\left\|\nabla f\left(x_k\right)\right\|^2 \\
& =f\left(x_k\right)-\frac{2\alpha - \alpha^2L}{2}\left\|\nabla f\left(x_k\right)\right\|^2
\end{aligned}
$$
When $\alpha <\frac{2}{L}$, the second term is always positive:
$$\frac{2\alpha - \alpha^2L}{2} > 0 $$
We have derived a bound on guaranteed progress.


Rewrite the equation:
$$\left\|\nabla f\left(x_{k}\right)\right\|^2 \leq \frac{2}{2\alpha - \alpha^2L} \left[f\left(x_{k}\right)-f\left(x_{k+1}\right)\right]$$
Sum up the squared norms of all the gradients up to iteration $T$:
$$
\sum_{k=0}^{T}\left\|\nabla f\left(x_{k}\right)\right\|^2 \leq \frac{2\alpha - \alpha^2L}{2} \sum_{k=0}^{T}\left[f\left(x_{k}\right)-f\left(x_{k+1}\right)\right]
$$
Apply some middle school algebra tricks:
$$
\begin{aligned}
\sum_{k=0}^{T}\left[f\left(x_{k}\right)-f\left(x_{k+1}\right)\right] & =f\left(x_0\right)-\underbrace{f\left(x_1\right)+f\left(x_1\right)}_0-\underbrace{f\left(x_2\right)+f\left(x_2\right)}_0-\ldots f\left(x_{T+1}\right) \\
& =f\left(x_0\right)-f\left(x_{T+1}\right)
\end{aligned}
$$
Now, we have:
$$
\begin{aligned}
\sum_{k=0}^{T}\left\|\nabla f\left(x_{k}\right)\right\|^2 &\leq \frac{2\alpha - \alpha^2L}{2} \left(f\left(x_0\right)-f\left(x_{T+1}\right)\right)\\
&\leq \frac{2\alpha - \alpha^2L}{2} \left(f\left(x_0\right)-f^*\right)\\
\end{aligned}
$$
Finally,
$$
\frac{\sum_{k=0}^{T}\left\|\nabla f\left(x_{k}\right)\right\|^2}{T} \leq \left[\frac{2\alpha - \alpha^2L}{2} \left(f\left(x_0\right)-f^*\right)\right]\cdot \frac{1}{T}
$$

This concludes the convergence of GD, with a convergence rate of $O(\frac{1}{T})$. Gradient descent requires $T=O(\frac{1}{\epsilon})$ iterations to achieve $\left\|\nabla f\left(x_T\right)\right\|^2 \leq \epsilon$.

## Summary of GD and Takeaways from the Proof

- Gradient descent is suitable for solving high-dimensional problems.
- Guaranteed progress if gradient is Lipschitz.
- Practical step size strategies (e.g. step decay step-size).


## Gradients of the Cost Function

Recall the vectorized form of our cost function:
$$A=\sigma\left(w^T X+b\right),$$
and
$$J=-\frac{1}{m} \sum \Big[Y * \log \left(A\right)+\left(1-Y\right) * \log \left(1-A\right)\Big],$$
where all operations (addition +, subtraction -, multiplication *) are point-wise, and the summation just sums over all the elements in the vector.

The partial derivatives are:
\begin{gathered}
\frac{\partial J}{\partial w}=\frac{1}{m} X(A-Y)^T \\
\frac{\partial J}{\partial b}=\frac{1}{m} \sum\left(A-Y\right),
\end{gathered}
